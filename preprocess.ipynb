```python
import pandas as pd
```

```python
# Load raw data
try:
    df = pd.read_csv('raw_bank_reviews.csv')
except FileNotFoundError:
    print("Error: raw_bank_reviews.csv not found. Run scraper.py first.")
    exit()
```

```python
# Preprocessing steps
# Problem 1: Remove duplicates
df.drop_duplicates(subset=['review', 'date', 'bank'], inplace=True)
print(f"Removed duplicates, {len(df)} reviews remain")
```

```python
# Problem 2: Handle missing data
df['review'].fillna('N/A', inplace=True)
df['date'].fillna('N/A', inplace=True)
df.dropna(subset=['rating'], inplace=True)  # Drop rows with missing ratings
print(f"After handling missing ratings, {len(df)} reviews remain")
```

```python
# Problem 3: Normalize date formats
# Convert to datetime, coerce invalid dates to NaT
df['date'] = pd.to_datetime(df['date'], errors='coerce')
# Format valid dates as YYYY-MM-DD, set invalid to 'N/A'
df['date'] = df['date'].apply(lambda x: x.strftime('%Y-%m-%d') if pd.notnull(x) else 'N/A')
print("Normalized dates to YYYY-MM-DD format")
```

```python
# Validate and normalize data
df['rating'] = df['rating'].astype(int)  # Ensure ratings are integers
df = df[df['rating'].between(1, 5)]  # Keep valid ratings (1–5)
df['source'] = 'Google Play'  # Ensure source is consistent
```

```python
# Save cleaned data
df.to_csv('clean_bank_reviews.csv', index=False)
print(f"Cleaned {len(df)} reviews and saved to clean_bank_reviews.csv")
```

```python
# Check missing data percentage
missing_percentage = df.isnull().mean() * 100
print("Missing data percentage:")
print(missing_percentage)
```

```python
# Validate total reviews
if len(df) < 1200:
    print(f"Warning: Only {len(df)} reviews after preprocessing, expected at least 1200")
```
