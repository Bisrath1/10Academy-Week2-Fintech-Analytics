```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df_sentiment = pd.DataFrame({
    'Bank': ['CBE', 'CBE', 'CBE', 'BOA', 'BOA', 'BOA', 'Dashen', 'Dashen', 'Dashen'],
    'Sentiment': ['Positive', 'Negative', 'Neutral']*3,
    'Percentage': [60, 25, 15, 30, 50, 20, 50, 30, 20]
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Bank', y='Percentage', hue='Sentiment', data=df_sentiment, palette=['#2ECC71', '#E74C3C', '#3498DB'])
plt.title('Sentiment Distribution by Bank')
plt.ylabel('Percentage (%)')
plt.xlabel('Bank')
plt.legend(title='Sentiment')
plt.grid(True)
plt.savefig('task_4/sentiment_bar.png')
plt.close()
```

```python
import matplotlib.pyplot as plt

labels = ['1 Star', '2 Stars', '3 Stars', '4 Stars', '5 Stars']
sizes = [40, 60, 60, 80, 160]  # From simulated data
colors = ['#E74C3C', '#F39C12', '#3498DB', '#2ECC71', '#1ABC9C']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('CBE Rating Distribution')
plt.axis('equal')
plt.savefig('task_4/cbe_rating_pie.png')
plt.close()
```

```python
df_themes = pd.DataFrame({
    'Theme': ['UI', 'Transaction Speed', 'Login Issues', 'Crashes', 'Customer Support'],
    'Percentage': [25, 20, 18, 15, 12]
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Theme', y='Percentage', data=df_themes, color='#2ECC71')
plt.title('Top Themes Across All Banks')
plt.ylabel('Percentage (%)')
plt.xlabel('Theme')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('task_4/theme_bar.png')
plt.close()
```

```python
df_trend = pd.DataFrame({
    'Month': ['Jan 2025', 'Feb 2025', 'Mar 2025', 'Apr 2025', 'May 2025', 'Jun 2025']*3,
    'Bank': ['CBE']*6 + ['BOA']*6 + ['Dashen']*6,
    'Sentiment_Score': [0.6, 0.62, 0.58, 0.6, 0.61, 0.59, 0.3, 0.28, 0.32, 0.29, 0.31, 0.3, 0.5, 0.51, 0.49, 0.5, 0.52, 0.51]
})

plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Sentiment_Score', hue='Bank', data=df_trend, palette=['#2ECC71', '#E74C3C', '#3498DB'])
plt.title('Sentiment Trends Over Time')
plt.ylabel('Average Sentiment Score')
plt.xlabel('Month')
plt.grid(True)
plt.savefig('task_4/sentiment_trend.png')
plt.close()
```

```python
from wordcloud import WordCloud

keywords = "crash login error slow transfer fast ui intuitive navigation reliable payment"
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(keywords)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Key Themes in Reviews')
plt.axis('off')
plt.savefig('task_4/keyword_cloud.png')
plt.close()
```
